```python
def load_data(file_path):
    S = pickle.load(open(file_path, 'rb'), encoding='latin1')#load pkl file as origin data

    S_activity = pd.DataFrame(S["activity"]).astype(int)
    S_activity.columns = ["Activity"]#add column name
    dic_activity = {1: "Sitting", 2: "Stairs", 3: "Soccer", 4: "Cycling", 5: "Driving", 6: "Lunch", 7: "Walking", 8: "Working"}#provide the dictionaty of activities

    
    S_label = pd.DataFrame(S['label'])#extract heart rates    
    S_label = pd.DataFrame(np.repeat(S_label.values,8,axis=0))#expand the length of label to be same as signal, multiply by 8 because sample rate is 8 seconds
    S_label.columns = ["Label"]
    if(np.size(S_label, axis = 0) < np.size(S_activity, axis = 0)):
        mean = S_label.mean()
        while(np.size(S_label, axis = 0) < np.size(S_activity, axis = 0)):
            S_label = S_label.append(mean, ignore_index=True)
    
    S_X = S['signal']#extract signal data
    S_X_chest = S_X['chest']
    S_X_wrist = S_X['wrist']#divide sigbals into chest and wrist

    chest_ACC = pd.DataFrame(S_X_chest['ACC'])
    chest_ACC = chest_ACC.iloc[::175, :]#calculate the mean per 175 of ACC
    chest_ACC.columns = ["ACC_x", "ACC_y", "ACC_z"]#add column names for 3-dimension data
    chest_ACC.reset_index(drop = True, inplace=True)

    chest_ECG = pd.DataFrame(S_X_chest['ECG'])
    chest_ECG = chest_ECG.iloc[::175, :]
    chest_ECG.columns = ["chest_ECG"]
    chest_ECG.reset_index(drop = True, inplace=True)

    chest_Resp = pd.DataFrame(S_X_chest['Resp'])
    chest_Resp = chest_Resp.iloc[::175, :]
    chest_Resp.columns = ["chest_Resp"]
    chest_Resp.reset_index(drop = True, inplace=True)

    #add columns of chest
    chest = pd.concat([chest_ACC], sort=False)
    chest["Resp"] = chest_Resp
    chest["ECG"] = chest_ECG
    chest.reset_index(drop=True, inplace=True)

    wrist_ACC = pd.DataFrame(S_X_wrist['ACC'])
    wrist_ACC = wrist_ACC.iloc[::8, :]
    wrist_ACC.columns = ["wrist_ACC_x", "wrist_ACC_y", "wrist_ACC_z"]
    wrist_ACC.reset_index(drop = True, inplace=True)

    wrist_EDA = pd.DataFrame(S_X_wrist['EDA'])
    wrist_EDA.columns = ["wrist_EDA"]

    wrist_BVP = pd.DataFrame(S_X_wrist['BVP'])
    wrist_BVP = wrist_BVP.iloc[::16, :]
    wrist_BVP.columns = ["wrist_BVP"]
    wrist_BVP.reset_index(drop = True, inplace=True)
 
    wrist_TEMP = pd.DataFrame(S_X_wrist['TEMP'])
    wrist_TEMP.columns = ["wrist_TEMP"]

    wrist = pd.concat([wrist_ACC], sort=False)
    wrist["BVP"] = wrist_BVP
    wrist["TEMP"] = wrist_TEMP
    wrist.reset_index(drop = True, inplace=True)

    signal = chest.join(wrist)#merge chest and wrist

    for k,v in S["questionnaire"].items() :
     signal[k] = v

    rpeaks = S['rpeaks']
    length = len(wrist_TEMP)
    counted_rpeaks = []
    index = 0 # index of rpeak element
    time = 175 # time portion
    count = 0 # number of rpeaks
    while(index < len(rpeaks)):
        rpeak = rpeaks[index]
    
        if(rpeak > time): # Rpeak appears after the time portion
            counted_rpeaks.append(count)
            count = 0
            time += 175      
        else:
            count += 1
            index += 1      
    # The Rpeaks are possiblely to be end before the time portion so the last portions should be filled as 0
    if(len(counted_rpeaks) < 36848):
        while(len(counted_rpeaks) < 36848):
            counted_rpeaks.append(0)
    
    peaks = pd.DataFrame(counted_rpeaks)
    peaks.columns = ["Rpeaks"]
    
    signal = signal.join(peaks)
    signal = signal.join(S_activity)
    signal = signal.join(S_label)
    signal['Subject'] = S["subject"]
    
    return signal

    
    
```
